```python
import matplotlib.pyplot as plt
#name of files
FILES_AES = ["type_AES__size_8.txt","type_AES__size_64.txt","type_AES__size_512.txt","type_AES__size_4096.txt","type_AES__size_32768.txt","type_AES__size_262144.txt","type_AES__size_2097152.txt"]
FILES_SHA = ["type_SHA__size_8.txt","type_SHA__size_64.txt","type_SHA__size_512.txt","type_SHA__size_4096.txt","type_SHA__size_32768.txt","type_SHA__size_262144.txt","type_SHA__size_2097152.txt"]
FILES_RSA = ["type_RSA__size_2.txt","type_RSA__size_4.txt","type_RSA__size_8.txt","type_RSA__size_16.txt","type_RSA__size_32.txt","type_RSA__size_64.txt","type_RSA__size_128.txt"]

#lists with data in microseconds
AES_same_files_encrypt = [93.639, 87.514, 84.286, 84.822, 85.244, 82.300, 81.789] 
AES_same_files_decrypt = [84.251, 79.143, 75.431, 82.778, 76.567, 73.578, 73.878]

AES_diff_files_encrypt = [82.633, 82.236, 82.148, 82.470, 82.052, 83.518, 82.016]
AES_diff_files_decrypt = [74.281, 73.193, 73.140, 73.711, 72.809, 73.736, 73.063]


RSA_same_files_encrypt = [324.214, 328.073, 324.314, 324.318, 324.310, 327.747, 326.861]
RSA_same_files_decrypt = [5526.320, 5555.318, 5546.135, 5527.488, 5552.108, 5548.235, 5519.604]

RSA_diff_files_encrypt = [330.783, 326.445, 325.968, 332.839, 328.324, 327.788, 328.696]
RSA_diff_files_decrypt = [5590.791, 5551.603, 5540.042, 5595.517, 5559.863, 5546.396, 5558.968]


SHA_same_files_encode = [19.418, 19.602, 21.592, 41.367, 178.789, 1307.833, 10417.233]
SHA_diff_files_encode = [20.415, 20.032, 22.383, 39.853, 187.746, 1329.407, 10433.946]

# bar_width
bar_width = 0.35
```

```python
# data for bar position on X axis
index = range(len(FILES_AES))

# plot
plt.bar(index, AES_same_files_encrypt, bar_width, label='Encryption',edgecolor='black')
plt.bar([i + bar_width for i in index], AES_same_files_decrypt, bar_width, label='Decryption', edgecolor='black')
plt.xlabel('File Index')
plt.ylabel('Time (microseconds)')
plt.title('AES Encryption/Decryption Times - same files')
plt.legend()
plt.xticks([i + bar_width / 2 for i in index], FILES_AES, rotation=45)
plt.tight_layout()
plt.show()
```

```python
# data for bar position on X axis
index = range(len(FILES_AES))

# plot
plt.bar(index, AES_diff_files_encrypt, bar_width, label='Encryption',edgecolor='black')
plt.bar([i + bar_width for i in index], AES_diff_files_decrypt, bar_width, label='Decryption', edgecolor='black')
plt.xlabel('File Index')
plt.ylabel('Time (microseconds)')
plt.title('AES Encryption/Decryption Times - different files')
plt.legend()
plt.xticks([i + bar_width / 2 for i in index], FILES_AES, rotation=45)
plt.tight_layout()
plt.show()
```

```python
# data for bar position on X axis
index = range(len(FILES_RSA))

# plot
plt.bar(index, RSA_same_files_encrypt, bar_width, label='Same files',edgecolor='black')
plt.bar([i + bar_width for i in index], RSA_diff_files_encrypt, bar_width, label='different files', edgecolor='black')
plt.xlabel('File Index')
plt.ylabel('Time (microseconds)')
plt.title('RSA Encryption Times - same vs different files')
plt.legend()
plt.xticks([i + bar_width / 2 for i in index], FILES_RSA, rotation=45)
plt.tight_layout()
plt.show()
```

```python
# data for bar position on X axis
index = range(len(FILES_RSA))

# plot
plt.bar(index, RSA_same_files_decrypt, bar_width, label='Same files',edgecolor='black')
plt.bar([i + bar_width for i in index], RSA_diff_files_decrypt, bar_width, label='different files', edgecolor='black')
plt.xlabel('File Index')
plt.ylabel('Time (microseconds)')
plt.title('RSA Decryption Times - same vs different files')
plt.legend()
plt.xticks([i + bar_width / 2 for i in index], FILES_RSA, rotation=45)
plt.tight_layout()
plt.show()
```

```python
# data for bar position on X axis
index = range(len(FILES_SHA))

# plot
plt.bar(index, SHA_same_files_encode, bar_width, label='Same files',edgecolor='black')
plt.bar([i + bar_width for i in index], SHA_diff_files_encode, bar_width, label='Different files', edgecolor='black')
plt.xlabel('File Index')
plt.ylabel('Time (microseconds)')
plt.title('SHA Digests Generation Times - same vs different files')
plt.legend()
plt.xticks([i + bar_width / 2 for i in index], FILES_SHA, rotation=45)
plt.tight_layout()
plt.show()
```
